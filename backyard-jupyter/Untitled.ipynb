```python
from pymisp import ExpandedPyMISP
# Adjust the import path for MISPtoSTIX21Parser based on your setup
from misp_stix_converter import MISPtoSTIX21Parser
from keys import misp_url, misp_key, misp_verifycert
import json

# Initialize MISP connection
misp = ExpandedPyMISP(misp_url, misp_key, misp_verifycert)

# Search parameters
misp_page = 1
misp_event_limit_per_page = 100
misp_event_filters = {}  # Adjust your filters here, if needed

# Function to convert STIX objects to JSON
def stix_to_json(stix_objects):
    # Assuming each STIX object has a serialize method
    stix_dicts = [json.loads(obj.serialize()) for obj in stix_objects]
    
    # Serialize the list of dictionaries to a JSON string
    stix_json = json.dumps(stix_dicts, indent=4)
    return stix_json

# Perform a search on MISP events
try:
    result = misp.search(controller='events', return_format='json', 
                         limit=misp_event_limit_per_page, page=misp_page, 
                         **misp_event_filters)
    if result:
        # Assuming result[0] exists and has data
        misp_event = result[0]  # Taking the first event for the example
        # Initialize the parser and convert the MISP event to STIX
        parser = MISPtoSTIX21Parser()
        parser.parse_misp_event(misp_event)
        stix_objects = parser.stix_objects
        
        # Convert and print the STIX objects to JSON
        readable_json = stix_to_json(stix_objects)
        print(readable_json)
    else:
        print("No events found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from pymisp import ExpandedPyMISP
from keys import misp_url, misp_key, misp_verifycert

def search_attributes_for_event(misp_url, misp_key, misp_verifycert, event_id):
    misp = ExpandedPyMISP(misp_url, misp_key, misp_verifycert)
    
    # Constructing the search query
    result = misp.search('attributes', eventid=event_id, to_ids=False, include_context=True)

    # Check and print the search results
    if result and 'Attribute' in result:
        for attribute in result['Attribute']:
            print(f"Attribute ID: {attribute['id']}, Type: {attribute['type']}, Value: {attribute['value']}")
    else:
        print(f"No attributes found for event ID: {event_id}")

# Specific event ID
event_id = '2667'
search_attributes_for_event(misp_url, misp_key, misp_verifycert, event_id)
```

```python
from pymisp import ExpandedPyMISP
from keys import misp_url, misp_key, misp_verifycert

def update_attribute_to_ids(misp_url, misp_key, misp_verifycert, event_id):
    misp = ExpandedPyMISP(misp_url, misp_key, misp_verifycert, debug=True)
    
    event = misp.get_event(event_id, pythonify=True)
    if event:
        for attribute in event.attributes:
            if not attribute.to_ids:
                # Directly update the to_ids property of the attribute
                attribute.to_ids = True
                
                try:
                    # Attempt to update the attribute
                    misp.update_attribute(attribute)
                    print(f"Updated Attribute ID: {attribute.id}, to_ids now set to True")
                except Exception as e:
                    print(f"Error updating attribute ID: {attribute.id}, Error: {e}")
    else:
        print(f"No data found for event ID: {event_id}")

# Specific event ID where you want to update attributes
event_id = '2667'
update_attribute_to_ids(misp_url, misp_key, misp_verifycert, event_id)
```
